```python
#Testes com as bibliotecas tabula-py e pypdf2

import PyPDF2 as pyf
from pathlib import Path

#Cria a pasta onde os PDFs serão salvos
Path("paginas").mkdir(exist_ok=True)

nome = "D:/João_Santana/Segundo Semestre/BackEnd2/Aula 02-02-2026/Arquivos pdf/primeiro.pdf"

#Abre o PDF corretamente
with open(nome,"rb") as arquivo:
    leitor = pyf.PdfReader(arquivo)

    for i,pagina in enumerate(leitor.pages):
        num_pagina = i+1

        escritor = pyf.PdfWriter()
        escritor.add_page(pagina)

        with open(f"paginas/Arquivo_{num_pagina}.pdf","wb") as novo_pdf:
            escritor.write(novo_pdf)
```
