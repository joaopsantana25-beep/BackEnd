```python

#Métodos Merge 

clientes_df = clientes_df[['ID Cliente', 'E-mail']]
cadastro_df = cadastro_df[["ID Produto","Nome do Produto"]]
lojas_df = lojas_df[["ID Loja", "Nome da Loja"]]

colunas_para_remover = ['Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10']
clientes_df = clientes_df.drop(columns=[coluna for coluna in colunas_para_remover if coluna in clientes_df.columns])

display(clientes_df)
```

```python
vendas_df = vendas_df.merge(cadastro_df,on='ID Produto')
vendas_df = vendas_df.merge(lojas_df,on='ID Loja')
vendas_df = vendas_df.merge(clientes_df,on='ID Cliente')

```
