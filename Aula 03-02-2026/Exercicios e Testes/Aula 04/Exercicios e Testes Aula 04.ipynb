```python
import PyPDF2 as pyf
from pathlib import Path

primeiro_arquivo = "D:/João_Santana/Segundo Semestre/BackEnd2/Aula 02-02-2026/Arquivos pdf/primeiro.pdf"
segundo_arquivo = "D:/João_Santana/Segundo Semestre/BackEnd2/Aula 02-02-2026/Arquivos pdf/segundo.pdf"

from pikepdf import Pdf, PdfImage 

arquivo = Pdf.open(primeiro_arquivo)

for pagina in arquivo.pages:
    for nome, imagem in pagina.images.items():
        imagem_salvar = PdfImage(imagem)
        imagem_salvar.extract_to(fileprefix = f"imagens/{nome}")
```
