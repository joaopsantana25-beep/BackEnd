Criação de um programa que gere senhas aleatórias para usuários 

```python
import numpy as nb
import pyautogui as gui
import string 


def gerar_codigo_aleatorio(tamanho):
    caracteres = string.ascii_letters + string.digits

    selecao = nb.random.choice(list(caracteres),size=tamanho)

    return ''.join(selecao)


while True:
    resposta = gui.prompt(text="Digite o tamanho da senha que você deseja: ",title="Página de senhas aleatórias")

    if not resposta:
        gui.alert("Você deixou em branco a página")
        continue

    if resposta.lower()=="sair":
        gui.alert("Terminando o código")
        break

    if not resposta.isdigit():
        gui.alert("Você digitou letras, por favor digite apenas números!")
        continue

    resposta = int(resposta)

    if resposta<=0:
        gui.alert("Por favor digite números positivos")
        continue

    else:
        gui.alert(text=f'A sua senha é: {gerar_codigo_aleatorio(resposta)}')

        resposta2=gui.prompt(text="Você deseja criar outra senha(S/N): ")

        resposta2=resposta2.upper()

        if not resposta2 or resposta2!='S':
            gui.alert(text="Encerrando o Programa")
            break




```

Teste de Automação com Pyautogui baseado na resposta do usuário

```python
import pyautogui as gui

while True:
    answer = gui.prompt("Qual rotina você deseja realizar (1,2,3): ")

    if not answer:
        gui.alert("Você deixou o arquivo em branco")
        continue
    
    if answer.lower()=="sair":
        gui.alert("Finalizando o Programa")
        break

    if not answer.isdigit():
        gui.alert("Por favor digite números")
        continue

    answer=int(answer)

    if answer<1 or answer>3:
        gui.alert("Digte um número dentro do intervalo dado: 1,2,3")
        continue

    gui.PAUSE=1

    if answer ==1:
        gui.press("win")
        gui.write("chrome")
        gui.press("enter")
        gui.hotkey('ctrl','l')
        gui.write("https://chatgpt.com/")
        gui.press("enter")
        gui.click(323, 570)
        gui.write("Como usar o gui para automatizar tarefas")
        gui.press("enter")

    if answer==2:
        gui.press("win")
        gui.write("chrome")
        gui.press("enter")
        gui.hotkey('ctrl','l')
        gui.write("https://youtu.be/JzS96auqau0?si=xk0OWV88F-OR2r0g")
        gui.press("enter")
    
    if answer ==3:
        gui.press("win")
        gui.write("chrome")
        gui.press("enter")
        gui.hotkey('ctrl','l')
        gui.write("https://email.com")
        gui.press("enter")

    gui.alert("Fim da Rotina")
   

    pergunta = gui.prompt("Você deseja fazer outra rotina (S/N): ")

    if not pergunta or pergunta.upper()!='S':
        gui.alert("Finalizando o Programa. Obrigado pelo uso!!!")
        break



    
```
