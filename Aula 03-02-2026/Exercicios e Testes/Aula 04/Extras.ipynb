```python
import numpy as nb
import pyautogui as gui
import string 


def gerar_codigo_aleatorio(tamanho):
    caracteres = string.ascii_letters + string.digits

    selecao = nb.random.choice(list(caracteres),size=tamanho)

    return ''.join(selecao)


while True:
    resposta = gui.prompt(text="Digite o tamanho da senha que você deseja: ",title="Página de senhas aleatórias")

    if not resposta:
        gui.alert("Você deixou em branco a página")
        continue

    if resposta.lower()=="sair":
        gui.alert("Terminando o código")
        break

    if not resposta.isdigit():
        gui.alert("Você digitou letras, por favor digite apenas números!")
        continue

    resposta = int(resposta)

    if resposta<=0:
        gui.alert("Por favor digite números positivos")
        continue

    else:
        gui.alert(text=f'A sua senha é: {gerar_codigo_aleatorio(resposta)}')

        resposta2=gui.prompt(text="Você deseja criar outra senha(S/N): ")

        resposta2=resposta2.upper()

        if not resposta2 or resposta2!='S':
            gui.alert(text="Encerrando o Programa")
            break




```
