```python
import pyautogui as gui
from PIL import Image

resposta = gui.prompt(text="Digite um número de 0 a 10: ",title="Caixa de Mensagem", default='')
bandeira = Image.open("D:/João_Santana/Segundo Semestre/BackEnd2/Aula 02-02-2026/Arquivos pdf/Bandeira.png")

if not resposta:
    print("Nenhum texto foi escrito")

else:
    gui.alert(text='Aqui diz que você é guay',title="Mensagem")
    bandeira.show()
    resposta=int(resposta)

    if resposta>10 or resposta<0:
        print("Por favor digite um valor válido entre 0 e 10")

    else:
        if resposta%2==0:
            print("Você digitou um número par")
        else:
            print("Você digitou um número ímpar")
```
