```python
import matplotlib.pyplot as plt

vendas_meses = [1500,1727,1350,999,1050,1027,1022,1500,2000,2362,2100,2762]
meses=["jan","fev","mar","abri","mai","jun","jul","ago","set","out","nov","dez"]

plt.plot(meses,vendas_meses)
plt.ylabel("Vendas")
plt.xlabel("Meses")
plt.axis([0,12,0,max(vendas_meses)+500])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

vendas = np.random.randint(10000,30000,50)
meses = np.arange(1,51)

plt.plot(meses,vendas,color="brown")
plt.ylabel("Vendas")
plt.xlabel("Meses")
plt.axis([0,len(meses)+5,0,max(vendas)+500])
plt.show()
```

```python
def padronizar_texto(texto):
    texto = texto.casefold()
    texto = texto.replace("  ",' ')
    texto = texto.strip()
    return texto
   


text= ["  ABC123","BSI23"]

text = list(map(padronizar_texto,text))
print(text)

```
