```python
'''Vers√£o M√°quina'''

import matplotlib.pyplot as plt

tipo = ["Conta de Luz", "Conta de √Ågua", "Streaming", "Carn√™s"]
gastos = [200, 250, 60, 490]
cores = plt.get_cmap("Dark2").colors

labels = [f"{t}\nR$ {v:.2f}" for t, v in zip(tipo, gastos)]

plt.figure(figsize=(5,5))
plt.pie(
    gastos,
    labels=labels,
    colors=cores,
    startangle=90,
    labeldistance=.50
)

plt.title("Gr√°fico de Gastos Mensais (R$)")
plt.show()
```

```python
'''
Crie um gr√°fico de linha mostrando a temperatura ao longo do dia.
'''

import numpy as nb
import matplotlib.pyplot as plt

T0=10


temperatura = []
horas = []

for i in range(24):

    temperatura.append(T0)
    horas.append(i)

    if i%2==0:
        T0+=2
    else:
        T0-=1

plt.plot(horas,temperatura)
plt.xlabel("Horas do Dia")
plt.ylabel("Temperatura (C¬∞)")
plt.title("Temperatura ao Longo do Dia")
plt.show()
```
