```python
import matplotlib.pyplot as plt

categorias = ['Eletr√¥nicos','Roupas','Alimentos','Livros']
quantidades = [45,30,15,10]
descontrole = [0.1,0,0,0]

cores = plt.get_cmap("Set1").colors

plt.pie(quantidades, labels=categorias,autopct='%.1f',colors=cores,explode=descontrole)
plt.title("Distribui√ß√£o de Categorias de Produtos em Porcentagem")
plt.show()
```

```python
import matplotlib.pyplot as plt

anos = [2018,2019,2020,2021]
vendas_online = [5,10,15,20]
vendas_loja = [10,8,6,4]

plt.bar(anos, vendas_online, label="Online")
plt.bar(anos,vendas_loja,bottom=vendas_online,label='Loja')

plt.xticks(ticks=anos, labels=anos)

plt.xlabel("Ano")
plt.ylabel('Vendas (10^6)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

meses = ['Jan','Fev','Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]

plt.annotate('Menor Valor',color='red',xy=(2,78),xytext=(3,80),arrowprops=dict(facecolor='red'))

plt.plot(meses,vendas,marker='o')
plt.xlabel("M√™s")
plt.ylabel("Vendas (10^6)")
plt.title("Crescientos das Vendas ao Longo do Ano")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

meses = ['Jan','Fev','Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]

marcadores = np.where(vendas == np.min(vendas),'x','')

for i, v in enumerate(vendas):
    if marcadores[i]=='x':
        plt.text(meses[i],v,marcadores[i],color='red',fontsize=20,ha='center')

plt.plot(meses,vendas)
plt.xlabel("M√™s")
plt.ylabel("Vendas (10^6)")
plt.title("Crescientos das Vendas ao Longo do Ano")
plt.grid(True)
plt.show()
```
