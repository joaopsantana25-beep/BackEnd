```python
import numpy as np
import matplotlib.pyplot as plt

#np.random.seed(42)
altura = np.random.normal(170,1,100)
peso = np.random.normal(70,1,100)

plt.scatter(altura,peso,color='green',s=20)
plt.title("Altura x Peso (1000 pessoas)")
plt.xlabel("Altura (Cm)")
plt.ylabel("Peso (Kg)")
plt.grid(True)
plt.show()
```
