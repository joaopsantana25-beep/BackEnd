```python
lista=[]

while True:
    
    numero = int(input("Digite um número de 1 a 10 (para sair digite 0): "))
    
    if numero==0:
        break
    
    if numero>10 or numero <0:
        print("Você digitou um número inválido")
        continue

    if numero%2==0:
        lista.append(numero)

if not lista:
    print("Você não digitou nenhum número par")

else:
    print(f'Você digitou {len(lista)} número pares')
```

```python
#Gerador de dados aleatórios para um teste

import numpy as np

#Gere dads de venas falsos para 30 dias
#Vamos supor que as vendas de um produto podem variar de 50 a 200 por dia

rng = np.random.default_rng(seed=42)
dados_vendas = rng.integers(low = 50, high=200, size=30)
print(dados_vendas)

#Operações para essas vendas 

media = np.round(np.mean(dados_vendas),2)
maximo = np.max(dados_vendas)
minimo = np.min(dados_vendas)
mediana = np.median(dados_vendas)
percentil = np.percentile(dados_vendas,50)
desvio_padrao = np.round(np.std(dados_vendas),2)
variancia = np.round(np.var(dados_vendas),2)

dia_max = np.argmax(dados_vendas)+1
dia_menos = np.argmin(dados_vendas)+1

print(f"O dia com mais vendas foi o dia {dia_max} com {maximo} vendas.")
print(f"O dia com menos vendas foi o dia {dia_menos} com {minimo} vendas.")
print(f'A média de vendas no mês foi {media}')
print(f"O valor da mediana das vendas é {mediana}")
print(f"O valor de percenti é {percentil}")
print(f"O desvio padrão é {desvio_padrao}")
print(f"A variancia é {variancia}")
```

```python
#Exercicios com salarios

import numpy as np

#Salários dos funcionários

salarios = np.array([3000,3500,4000,2000,4500,4000,5000])

#Calcular a média salarial

media_salarial = np.round(np.mean(salarios),2)
print(f"A média salarial é R$ {media_salarial}")

#Identificar salarios acima da media 
funcionarios_acima_media = np.where(salarios>media_salarial)[0]
print(f"Os funcionários com salários acima da média são {funcionarios_acima_media}")
print(f"Os salarios acima da media são {salarios[funcionarios_acima_media]} ")

#Dizer se o salario é acima ou abaixo da media 
print(np.where(salarios>media_salarial,'Acima da média','Abaixo da média'))

#Salarios bônus
salarios_bonus=np.where(salarios<media_salarial,salarios*1.1,salarios)
print(salarios)
print(salarios_bonus)

#Salarios na faixa de 3000 e 4500
print(np.where((salarios>=3000) & (salarios<=4500)))

#Reajuste salarial de 5% para esses que estão dentro dessa faixa 
print(salarios)
print(np.where((salarios>=3000) & (salarios<=4500),salarios*1.05,salarios))

#Reajuste salarial de 10% para pessoas abaixo de 3000 ou acima de 4500
print(salarios)
salarios_ajustados = np.where((salarios<3000) | (salarios>4500),salarios*1.1,salarios)
print(salarios_ajustados)
```

```python
#Reshape de um array

import numpy as np

#Vendas diarias para 2 semanas

vendas = np.array ([200,220,250,210,300,280,230,210,220,240,230,210,280,220])

#Reorganizar os dados em uma matriz de 2x7

vendas_reshaped = np.reshape(vendas,(2,7))
print(vendas_reshaped)

#Operações com o formato da matriz
print(vendas.ndim)
print(vendas.shape)

print(vendas_reshaped.ndim)
print(vendas_reshaped.shape)

#Printando cada matriz de venda separadamente
print(vendas_reshaped[0])

print(vendas_reshaped[1])

#Soma de cada array

print(vendas_reshaped.sum())

print(vendas_reshaped.sum(axis=1))
print(vendas_reshaped.mean(axis=1))

```
